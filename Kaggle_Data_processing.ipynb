## Import required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn import preprocessing
```

## Import train and test data sets

Here you have to download train.csv and test.csv files from https://www.kaggle.com/c/house-prices-advanced-regression-techniques/data location and upload them into Google colab runtime.

```python
train_df = pd.read_csv('/content/train.csv')
test_df = pd.read_csv('/content/train.csv')
train_df.info()
test_df.info()
```

# Remove null values in dataset


## Retrive columns with null values

```python
columns_with_null_values = list(train_df.isnull().sum().items())
columns_with_null_values
```

As per the results given in previous code,  some feature columns contain null values.  Removed  columns which have more than 70% null values.

```python
train_df = train_df.drop(['Alley', 'FireplaceQu','PoolQC','Fence','MiscFeature', 'Unnamed: 81'], axis=1)
```

```python
test_df = test_df.drop(['Alley', 'FireplaceQu','PoolQC','Fence','MiscFeature'], axis=1)
```

# Encoding

## Select categorical variables

```python
train_df.describe(include='O')
```

```python
test_df.describe(include='O')
```

```python
labelencoder = LabelEncoder()
le = preprocessing.LabelEncoder()
```

## Encode categorical variables in train data set

```python
objFeatures = train_df.select_dtypes(include="object").columns

for feat in objFeatures:
    train_df[feat] = le.fit_transform(train_df[feat].astype(str))
```

```python
train_df.info()
```

## Encode categorical variables in test data set

```python
objFeatures = test_df.select_dtypes(include="object").columns

for feat in objFeatures:
    test_df[feat] = le.fit_transform(test_df[feat].astype(str))
```

```python
test_df.info()
```

# Data Normalization

## Train data normalization

```python
def train_normalize(feature_name):
  max_value = train_df[feature_name].max()
  print(feature_name,max_value)
  min_value = train_df[feature_name].min()
  print(min_value)
  train_df[feature_name] = (train_df[feature_name] - min_value) / (max_value - min_value)
  print( train_df[feature_name].max())
  print(feature_name, train_df[feature_name].min())
  return train_df[feature_name]
```

```python
normalized_features = ['2ndFlrSF','EnclosedPorch','3SsnPorch','ScreenPorch','MiscVal','MasVnrArea','BsmtFinSF1','BsmtFinSF2']

for feature_name in normalized_features:
    train_normalize(feature_name)
```

## Test data normalization

```python
def test_normalize(feature_name):
  max_value = test_df[feature_name].max()
  print(feature_name,max_value)
  min_value = test_df[feature_name].min()
  print(min_value)
  test_df[feature_name] = (test_df[feature_name] - min_value) / (max_value - min_value)
  print( test_df[feature_name].max())
  print(feature_name, test_df[feature_name].min())
  return test_df[feature_name]
```

```python
normalized_features = ['2ndFlrSF','EnclosedPorch','3SsnPorch','ScreenPorch','MiscVal','MasVnrArea','BsmtFinSF1','BsmtFinSF2']

for feature_name in normalized_features:
    test_normalize(feature_name)
```
