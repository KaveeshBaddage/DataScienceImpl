```python
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger') 

text = "learn php from guru99"
tokens = nltk.word_tokenize(text)
print(tokens)
nltk.pos_tag(text)
```

```python
nltk.pos_tag(tokens)
#print("The type is : ",type(text))
```
